```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import librosa
import librosa.display
from IPython.display import Audio
```

```python
dip_1, sr = np.array(librosa.load('1/BR1_belt.wav'))
dip_2, sr = np.array(librosa.load('1/BL2_belt.wav'))
ayu_1, sr = np.array(librosa.load('2/BR1_belt.wav'))               #SAME SAMPLING FREQUENCIES.
ayu_2, sr = np.array(librosa.load('2/BL2_belt.wav'))
```

```python
plt.figure(figsize=(12,6))
plt.grid()
librosa.display.waveplot(dip_1, sr=sr)
plt.title("Dipankar's R1")
plt.xlabel('Time')
plt.ylabel('Pressure')
plt.show()
```
